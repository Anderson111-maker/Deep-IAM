```python
import pandas as pd
df = pd.read_csv('./1668008312256-AR6_Scenarios_Database_World_v1.1.csv/AR6_Scenarios_Database_World_v1.1.csv')
meta = pd.read_excel('./1668008312256-AR6_Scenarios_Database_World_v1.1.csv/AR6_Scenarios_Database_metadata_indicators_v1.1.xlsx',sheet_name = 'meta')
```

```python
meta.head()
```

```python
year_columns = [str(x) for x in list(range(2005,2101,5))] # At five-year intervals, take the values of the variables for the period 2005-2100
variable = 'AR6 climate diagnostics|Infilled|Emissions|Kyoto Gases (AR6-GWP100)'
df_sub = df.loc[df.Variable == variable,['Model','Scenario']+year_columns]#"df_sub" contains the values of the variables during 2005-2100 and their corresponding Models and Scenarios
df_sub.head()
```

```python
data = df_sub.merge(meta.loc[:,['Model','Scenario','Category','Category_name']], on=['Model', 'Scenario'])#Extract the category corresponding to each scenario in the meta and merge it with df_sub
```

## Imputation

```python
data.isna().sum()#The proportion of missing data at each point in time
```

```python
columns_drop = [str(x) for x in range(2005,2100,10)]  #Only data for the period 2010-2100 will be retained (10-year interval).
data_input = data.drop(columns = columns_drop)
data_input.sample()
```

```python
data_input.isna().sum()/data_input.shape[0]
```

```python
#Fill in the missing values with the mean of the corresponding year.
data_input[data_input.select_dtypes(include=['number']).columns] = data_input.select_dtypes(include=['number']).fillna( data_input.select_dtypes(include=['number']).mean())
```
